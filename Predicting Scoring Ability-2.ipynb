```python
combined = pd.merge(players, seasons_stats, on='Player')
combined = pd.merge(combined,player_data, on='Player')
combined.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y'], inplace=True) #Dropping unneeded columns

combined['FG%'].fillna(combined['FG%'].mean(), inplace=True)
combined['FT%'].fillna(combined['FT%'].mean(), inplace=True)

combined.head()
```

Something we need to take of is the fact that some of FG% and FT% are NaN. This is because the player never had a chance to shoot that kind of shot. Because of this, we will replace those values with the mean of the column to avoid any errors later on. In addition to this, some FG% and FT% are 1, meaning they never missed. That's likely due to the fact that they took a low number of shots. I will not be modifying this, but this is likely going to affect our results later on.

```python
sns.scatterplot(data=combined, x="weight_kg", y="FG%", hue='position')
plt.title('Field Goal and Weight Distribution of each Position')
plt.show()
```

<!-- With this, we can see the positions are duplicated and needs some cleaning. What I mean by this, the position SF-SG is the same as SG-SF.  -->

Notice that some positions are duplicated. G-F, or Guard-Forward, is the same as F-G. The same can be said about F-C and C-F. Because of this, we will need to fix this in our graph. In addition, the graph is a bit hard to distinguish between all positions. Because of this, we will make a graph for each position instead!

```python
for index, row in combined.iterrows():
    currentPosition = row['position']
    
    if (currentPosition == 'G-F'):
        combined.at[index, 'position'] = 'F-G'
    elif (currentPosition == 'F-C'):
        combined.at[index, 'position'] = 'C-F'

combined.position.unique()
```

We have now fixed the duplicates in the positions, now let's create a graph for each position. Let's look at the FG% and FT% compared to the weight and then height, for each position!

<h2>Predicting Shooting Ability from Physical Characteristics</h2>

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

x = combined[['height_cm','weight_kg', 'FTA', 'FGA']]
y = combined['FG%']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

LR = LinearRegression()
LR.fit(x_train,y_train)

y_prediction = LR.predict(x_test)
y_prediction

score=r2_score(y_test,y_prediction)
print('r-squared score is ',score)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

x = combined[['height_cm','weight_kg', 'FTA', 'FGA']]
y = combined['FT%']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

LR = LinearRegression()
LR.fit(x_train,y_train)

y_prediction = LR.predict(x_test)
y_prediction

score=r2_score(y_test,y_prediction)
print('r-squared score is ',score)
```

This r-squared score is fairly low. This is likely due to the fact that there were a couple outliers. More specifcially, there were some players with 100% shooting accuracy or 0%. These players likely didn't take many shots. Despite putting into our model, the number of free throw attempts and field goal attempts, this still resulted in a low r-squared values.
